```python
SELECT
    customer_name,
    ROUND(
        SUM(CASE WHEN order_type = 'buy' THEN order_amount * 0.001 ELSE order_amount * 0.0015 END),
        2
    ) AS total_fees_paid
FROM
    orders
GROUP BY
    customer_name
ORDER BY
    customer_name;
```

```python
SELECT customers.customer_name,
    ROUND(
        SUM(CASE WHEN order_type = 'buy' THEN order_amount * 0.001 ELSE order_amount * 0.0015 END),
        2
    ) AS total_fees_paid
FROM
    customer
JOIN orders
ON   orders.customer_id = customer.id
GROUP BY
    customer_name
ORDER BY
    customer_name;
```

```python
SELECT

  
    CONCAT(c.first_name, ' ', c.last_name) AS customer,
    GROUP_CONCAT(DISTINCT cam.name ORDER BY cam.name ASC) AS campaign,
    COUNT(e.id) AS total
FROM
    customer c
JOIN
    campaign cam ON c.id = cam.customer_id
JOIN
    events e ON cam.id = e.campaign_id 

#AND e.status = 'success'

GROUP BY
    customer
ORDER BY
    total DESC
LIMIT 2

UNION

SELECT
    'failure' AS type,
    CONCAT(c.first_name, ' ', c.last_name) AS customer,
    GROUP_CONCAT(DISTINCT cam.name ORDER BY cam.name ASC) AS campaign,
    COUNT(e.id) AS total
FROM
    customer c
JOIN
    campaign cam ON c.id = cam.customer_id
JOIN
    events e ON cam.id = e.campaign_id AND e.status = 'failure'
GROUP BY
    customer
ORDER BY
    total DESC
LIMIT 2;
```
